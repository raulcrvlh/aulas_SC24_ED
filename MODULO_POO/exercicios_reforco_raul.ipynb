# Exercício 1

Classe Pessoa: Crie uma classe que modele uma pessoa:

Atributos: nome, idade, peso e altura


Métodos: Envelhercer, engordar, emagrecer, crescer. 

Obs: Por padrão, a cada ano que nossa pessoa envelhece, sendo a idade dela menor que 21 anos, ela deve crescer 0,5 cm.

```python
from datetime import datetime as dt

class Pessoa:
    def __init__(self, nome:str, ano_nascimento:int, peso:float, altura:float):
        self._nome = nome
        self._ano_nascimento = ano_nascimento
        self._ano_atual = int(dt.now().strftime('%Y'))
        self._idade = self.ano_atual - self.ano_nascimento
        self._peso = peso
        self._altura = altura
        self._idade_anterior = self.idade # salva idade anterior como idade inicial

    def __repr__(self) -> str:
        pessoa = f'Nome: {self.nome} \nIdade: {self.idade} \nPeso: {self.peso} \nAltura: {self.altura:.2f}'
        return pessoa

    def envelhecer(self) -> int:
        self.idade += 1 # aumenta um ano na idade
    
    def crescer(self) -> float:
        if self.idade < 21: # verifica se a idade atual é menor que 21 anos
            if self.idade > self.idade_anterior: # verifica se a idade atual é maior do que a idade anterior, para ter certeza que envelheceu
                envelheceu = self.idade - self.idade_anterior # verifica quantos anos a pessoa envelheceu
                self.altura += envelheceu * 0.05 # quantidade de anos que envelheceu * 0.05
                self.idade_anterior = self.idade # salva idade anterior como nova idade atual
                return self.altura
            else:
                return self.altura
        else:
            print('Idade máxima de crescimento atingida.')

    
    def engordar(self) -> float:
        # para cada ano aumenta 2kg no peso até os 21 anos, depois verifica quantos kg engordou
        if self.idade < 21:
            if self.idade > self.idade_anterior: # verifica se a idade atual é maior do que a idade anterior, para ter certeza que envelheceu
                envelheceu = self.idade - self.idade_anterior # verifica quantos anos a pessoa envelheceu
                self.peso += envelheceu * 2.0 # quantidade de anos que envelheceu * 2.0kg por ano
                # self.idade_anterior = self.idade # salva idade anterior como nova idade atual
                return self.peso
            else:
                return self.peso
        else:
            print('Quantos kg engordou?')
            qtd_quilos = float(input('Quantos kg engordou? Ex: 1.5'))
            self.peso += qtd_quilos
            return self.peso

    def emagrecer(self, quilos:float) -> float:
        self.peso -= quilos
        return self.peso
    
    @property
    def nome(self):
        return self._nome
    
    @property
    def ano_nascimento(self):
        return self._ano_nascimento
    
    @property
    def ano_atual(self):
        return self._ano_atual
    
    @property
    def idade(self):
        return self._idade
    
    @property
    def idade_anterior(self):
        return self._idade_anterior
    
    @property
    def altura(self):
        return self._altura
    
    @property
    def peso(self):
        return self._peso
```

Classe Conta Corrente: Crie uma classe para implementar uma conta corrente. 

A classe deve possuir os seguintes atributos: número da conta, nome do correntista e saldo. 

Os métodos são os seguintes: alterarNome, depósito e saque; No construtor, saldo é opcional, com valor default zero e os demais atributos são obrigatórios.

```python
class ContaCorrente:
    def __init__(self, num_conta:str, nome_correntista:str, saldo:float = 0) -> None:
        self.num_conta = num_conta
        self.nome_correntista = nome_correntista
        self.saldo = saldo

    def __repr__(self) -> str:
        infos = f'Número da Conta: {self.num_conta} \nNome do Correntista: {self.nome_correntista} \nSaldo em conta: {self.saldo}'
        return infos 

    def alterarNome(self, novo_nome:str) -> str:
        self.nome_correntista = novo_nome
        return self.nome_correntista
    
    def deposito(self, valor:float) -> float:
        self.saldo += valor
        return self.saldo
    
    def saque(self, valor: float) -> float:
        if self.saldo > 0 and valor < self.saldo:
            self.saldo -= valor
        else:
            return f'Saldo insuficiente, você pode sacar no máximo R${self.saldo}'
        return self.saldo
```

```python
class Cliente:
    def __init__(self, num_conta:str, nome_correntista:str, saldo:float = 0) -> None:
        self.num_conta = num_conta
        self.nome_correntista = nome_correntista
        self.saldo = saldo

    def mostrar_cliente(self) -> dict:
        return {"Número da conta": self.num_conta, "Cliente": self.nome_correntista, "Saldo": self.saldo}

class ContaCorrente:
    def __init__(self) -> None:
        self.cadastro = {}

    def cadastrar_cliente(self, cpf:str):
        nome = input("Digite o nome do cliente: ")
        numero_conta = input("Digite o número da conta: ")
        saldo = input("Digite o saldo do cliente: ")
        self.cadastro[cpf] = Cliente(num_conta=numero_conta, nome_correntista=nome, saldo=saldo) # cria os dados do cliente na chave cpf 
        print(f"Cliente com CPF {cpf} cadastrado com sucesso!")

    def alterar_nome(self, cpf: str):
        if cpf in self.cadastro: # se cpf está consta no cadastro de clientes
            saldo_atual = self.cadastro[cpf].saldo
            nome = input("Digite o novo nome do cliente: ") # verifica novo nome
            self.cadastro[cpf] = Cliente(self.cadastro[cpf].num_conta, nome_correntista=nome, saldo=saldo_atual) # altera o nome no cadastro
            print(f'Cadastro do cliente com CPF {cpf} alterado com sucesso!')
            print(f"Nome Atualizado: {self.cadastro[cpf].nome_correntista}")
        else:
            print("Cliente não encontrado.")

    def registar_deposito(self, cpf:str, valor:float): # registra um depósito com base na chave cpf
        if cpf in self.cadastro: # verifica se o cpf consta no dicionário de cadastro
            saldo_atual = float(self.cadastro[cpf].saldo) # recebe saldo altual da conta
            num_conta = self.cadastro[cpf].num_conta # recebe numero da conta
            cliente = self.cadastro[cpf].nome_correntista # recebe nome do cliente

            saldo_atualizado = saldo_atual + valor # atualiza saldo somando o valor depositado
            self.cadastro[cpf] = Cliente(num_conta, cliente, saldo=saldo_atualizado) # atualiza dados do cliente
            print(f'Saldo do cliente com CPF {cpf} alterado com sucesso.')
            print(f"Saldo Atualizado: {self.cadastro[cpf].saldo}")
        else:
            print("Cliente não encontrado.")

    def registar_saque(self, cpf:str, valor:float):# registra um saque com base na chave cpf
        if cpf in self.cadastro: # verifica se o cpf consta no dicionário de cadastro
            saldo_atual = float(self.cadastro[cpf].saldo) # recebe saldo atual
            num_conta = self.cadastro[cpf].num_conta # recebe numero da conta
            cliente = self.cadastro[cpf].nome_correntista # recebe nome do cliente
            
            if valor <= saldo_atual: # verifica se o valor que será sacado é menor ou igual ao saldo atual da conta
                saldo_atualizado = saldo_atual - valor # atualiza saldo subtraindo o valor existente da conta pelo valor do saque
                self.cadastro[cpf] = Cliente(num_conta, cliente, saldo=saldo_atualizado) # atualiza dados do cliente
                print(f'Saldo do cliente com CPF {cpf} alterado com sucesso.')
                print(f"Saldo Atualizado: {self.cadastro[cpf].saldo}")
            else:
                print(f'Saldo insuficiente. O valor disponível para saque é {saldo_atual}') # se o valor que deseja sacar for maior do que o valor que consta no saldo, retorna uma mensagem de alerta.

    def imprimir_cadastros(self): # imprime todos os cadastros existentes
        if not self.cadastro: # se a lista de cadastro estiver vazia
            print("Nenhum cliente cadastrado.") # imprime mensagem de alerta
        else:
            for cpf, cliente in self.cadastro.items(): # para cada cpf e cliente no dicionário de cadastros
                print(f"CPF: {cpf}") # imprime cpf
                info_cliente = cliente.mostrar_cliente() # recebe informações do cliente na chave cpf
                for chave, valor in info_cliente.items(): # para cada chave valor do dicionário de informações do cliente
                    print(f"{chave}: {valor}") # imprime chave e valor
                print("-" * 20)

    def sair(self) -> bool:
        print("Saindo do sistema. Até logo.")
        return True
    
    def menu(self) -> None:
        while True:
            print("-" * 45)
            print("- Escolha uma opção:                       -")
            print("- 1. Cadastrar nova conta corrente         -")
            print("- 2. Alterar cadastro existente            -")
            print("- 3. Registar depósito                     -")
            print("- 4. Registrar saque                       -")
            print("- 5. Imprimir clientes cadastrados         -")
            print("- 6. Sair                                  -")
            print("-" * 45)

            opcao = input("Opcao: ")

            if opcao == "1":
                cpf = input("Digite o CPF do cliente: ")
                if cpf not in self.cadastro:
                    self.cadastrar_cliente(cpf)
                else:
                    print("Cliente já cadastrado.")

            elif opcao == "2":
                cpf = input("Digite o CPF do cliente que deseja alterar: ")
                self.alterar_nome(cpf)

            elif opcao == "3":
                cpf = input("Digite o CPF para quem será feito o depósito: ")
                deposito = float(input("Digite o valor que será depositado: "))
                self.registar_deposito(cpf, deposito)

            elif opcao == "4":
                cpf = input("Digite o CPF da conta que será feito o saque: ")
                saque = float(input("Digite o valor que deseja sacar: "))
                self.registar_saque(cpf, saque)

            elif opcao == "5":
                self.imprimir_cadastros()

            elif opcao == "6":
                if self.sair():
                    break

            else:
                print("Opção inválida. Tente novamente.")

sistema_CC = ContaCorrente()
sistema_CC.menu()
```

# Exercício 3

Classe Bichinho Virtual:Crie uma classe que modele um Tamagushi (Bichinho Eletrônico):

Atributos: Nome, Fome, Saúde e Idade b. Métodos: Alterar Nome, Fome, Saúde e Idade; Retornar Nome, Fome, Saúde e Idade 

Obs: Existe mais uma informação que devemos levar em consideração, o Humor do nosso tamagushi, este humor é uma combinação entre os atributos Fome e Saúde, ou seja, um campo calculado, então não devemos criar um atributo para armazenar esta informação por que ela pode ser calculada a qualquer momento.


### Regras de Humor do Tamagushi:

- Fome = 0 -> sem fome
- Fome = 10 -> com muita fome

- Saúde = 0 -> péssima
- Saúde = 10 -> excelente

Humor Excelente:
- Fome ≤ 3 e Saúde ≥ 8.
- O Tamagushi estará muito feliz e brincalhão.

Humor Bom:
- Fome ≤ 5 e Saúde ≥ 6.
- O Tamagushi estará contente, mas talvez precise de alguma atenção.

Humor Neutro:
- Fome ≤ 7 e Saúde ≥ 4.
- O Tamagushi estará tranquilo, mas não tão ativo. Pode estar começando a sentir fome ou ter pequenos problemas de saúde.

Humor Ruim:
- Fome > 7 ou Saúde ≤ 3.
- O Tamagushi estará irritado ou apático. Ele precisa urgentemente de comida ou cuidados de saúde.

Humor Péssimo:
- Fome = 10 ou Saúde ≤ 2.
- O Tamagushi estará muito mal, possivelmente doente e extremamente faminto. A atenção imediata é necessária para evitar que sua condição piore.

Atualizações com o Tempo:
- Fome: A fome aumenta ao longo do tempo. A cada X minutos, o nível de fome aumenta em 1 ponto, até atingir 10.
- Saúde: A saúde diminui ao longo do tempo se a fome estiver alta (acima de 7). A cada X minutos, se a fome for maior que 7, a saúde diminui em 1 ponto.

```python
from datetime import datetime as dt

class BichoVirtual:
    def __init__(self, nome:str, fome:float = 5.0, saude:float = 5.0, idade:int = 0):
        self.nome = nome
        self.fome = fome
        self.saude = saude
        self.idade = idade
        self.humor = self.calcular_humor()

    def calcular_humor(self):
        if self.fome <= 3 and self.saude >= 8:
            return 'EXCELENTE'
        elif self.fome <= 5 and self.saude >= 6:
            return 'BOM'
        elif self.fome <= 7 and self.saude >= 4:
            return 'TRANQUILO'
        elif self.fome > 7 or self.saude <= 3:
            return 'RUIM'
        elif self.fome == 10 or self.saude <= 2:
            return 'PÉSSIMO'
        else:
            return 'Desconhecido'

    def mostrar_infos(self):
        return {
            "Nome": self.nome,
            "Fome": round(self.fome,2), # arredonda os valroes float para ficar com no máximo 2 casas decimais
            "Saude": round(self.saude,2),
            "Idade": self.idade,
            "Humor": self.humor
        }
    
class Controle:
    def __init__(self) -> None:
        self.minuto = dt.now().minute # recebe o valor do minuto inicial do jogo com base na hora do computador
        self.bicho = None

    def criar_tamagushi(self):
        nome = input("Digite o nome do seu bichinho virtual: ") # para criar o bichinho virtual é necessário colocar apenas o nome
        self.bicho = BichoVirtual(nome) # cria um dicionário com os dados do bichinho virtual {nome, fome=5, saude=5, idade=0}

    def envelhecer(self, tempo:int) -> None:
        envelheceu = tempo - self.minuto # calcula o quanto o tamagushi envelheceu (envelhece 1 ano por minuto)
        if envelheceu > 0:
            self.bicho.idade += envelheceu # incrementa a idade 
            self.bicho.saude -= 1.0 # sempre que ganha um ano, diminui um nível da saúde
            self.bicho.fome += 1.0 # sempre que ganha um ano, aumenta um nível da fome 
            self.minuto = tempo # atualiza minuto inicial para novos cálculos


    def alimentar(self):
        if self.bicho.fome > 0 and self.bicho.fome <= 10: # se a fome estiver entre 0 e 10
            self.bicho.fome -= 1 # diminui o nível da fome
        else:
            print("O bichinho já está sem fome!") # caso contrário, diz que ele já está sem fome

    def brincar(self):
        if self.bicho.saude > 0 and self.bicho.saude <= 10: # se a saúde estiver entre 0 e 10
            self.bicho.saude += 1.0 # aumenta o nível de saúde
        if self.bicho.fome > 0 and self.bicho.fome <= 10: 
            self.bicho.fome += 1.0 # toda vez que brincar, aumenta o nível de fome

    def nivelar_fome(self, tempo:int):
        fome_incremento = (tempo - self.minuto) # calcula o incremento da fome com base em quanto tempo passou
        if self.bicho.fome < 0:
            self.bicho.fome = 0.0 # se fome menor 0, retorna 0
        elif self.bicho.fome > 10.0:
            self.bicho.fome = 10 # se fome maior que 10, retorna 10
        else:
            self.bicho.fome += fome_incremento #se fome está entre o range (0 a 10), incrementa o nível de fome com o passar do tempo

    def nivelar_saude(self, tempo:int):
        saude_decremento = (tempo - self.minuto) # calcula o nível de saúde com base em quanto tempo passou
        if self.bicho.saude < 0:
            self.bicho.saude = 0.0 # se saúde menor que 0, retorna 0
        if self.bicho.saude > 10:
            self.bicho.saude = 10.0 # se saúde maior que 10, retorna 0
        self.bicho.saude -= saude_decremento # decrementa o nível da saúde com o passar do tempo

    def atualizar_humor(self):
        self.bicho.humor = self.bicho.calcular_humor() # atualiza o nivel do humor com base nos cálculos
        print(f'O {self.bicho.nome} está com humor: {self.bicho.humor}')

    def sair(self) -> None:
        print("Você está saindo do programa. Até logo.")
        return True
    
    def menu(self):
        while True:
            if self.bicho:
                self.minuto_atual = dt.now().minute
                self.envelhecer(self.minuto_atual)
                self.nivelar_fome(self.minuto_atual)
                self.nivelar_saude(self.minuto_atual)
                self.atualizar_humor()
                print(self.bicho.mostrar_infos())

            print("\nEscolha uma opção: ")
            print("1. Criar Bichinho Virtual")
            print("2. Alimentar")
            print("3. Brincar")
            print("4. Verificar humor")
            print("5. Sair")

            opcao = input("Escolha uma opção: ")

            if opcao == "1":
                self.criar_tamagushi()

            elif opcao == "2" and self.bicho:
                self.alimentar()

            elif opcao == "3" and self.bicho:
                self.brincar()

            elif opcao == "4" and self.bicho:
                self.atualizar_humor()

            elif opcao == "5":
                if self.sair():
                    break
            else:
                print("Opção inválida ou bichinho não criado ainda.")

# Inicializando o jogo
tamagushi = Controle()
tamagushi.menu()
```
